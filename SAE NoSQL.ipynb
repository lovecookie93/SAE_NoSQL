##### 2.Pour chaque employé, le nombre de clients, le nombre de commandes et le montant total de celles-ci ;


```python
pandas.read_sql_query(
"""
SELECT 
    e.employeeNumber,
    e.firstName,
    e.lastName,
    COUNT(DISTINCT c.customerNumber) AS numberOfCustomers,
    COUNT(DISTINCT o.orderNumber) AS numberOfOrders,
    SUM(p.amount) as TotalOrderAmount
FROM Employees e 
LEFT JOIN Customers c ON e.employeeNumber = c.salesRepEmployeeNumber
LEFT JOIN Orders o ON c.customerNumber = o.customerNumber
LEFT JOIN Payments p ON o.customerNumber = p.customerNumber
GROUP BY e.employeeNumber, e.firstName, e.lastName;
""",
conn)
```

##### 3.Idem pour chaque bureau (nombre de clients, nombre de commandes et montant total), avec en plus le nombre de clients d’un pays différent, s’il y en a ;

```python
pandas.read_sql_query(
"""
SELECT 
    Offices.officeCode,
    Offices.city,
    Offices.country,
    COUNT(DISTINCT Customers.customerNumber) AS nombre_clients,
    COUNT(DISTINCT Orders.orderNumber) AS nombre_commandes,
    SUM(p.amount) AS montant_total_commandes,
    count(distinct  CASE 
            WHEN Customers.country != Offices.country THEN 1
            ELSE NULL
        END) AS clients_differents_pays
FROM 
    Offices
LEFT JOIN Employees 
    ON Offices.officeCode = Employees.officeCode
LEFT JOIN Customers 
    ON Employees.employeeNumber = Customers.SalesRepEmployeeNumber
LEFT JOIN Orders 
    ON Customers.customerNumber = Orders.customerNumber
LEFT JOIN Payments p ON Orders.customerNumber = p.customerNumber
GROUP BY 
    Offices.officeCode, Offices.city, Offices.country
;
""",
conn)
```

##### 4.Pour chaque produit, donner le nombre de commandes, la quantité totale commandée, et le nombre de clients différents ;


```python
pandas.read_sql_query(
"""
SELECT Products.productCode, Products.productName, COUNT(OrderDetails.orderNumber), SUM(OrderDetails.quantityOrdered), COUNT(DISTINCT(Orders.customerNumber))
FROM Products 
LEFT JOIN OrderDetails on OrderDetails.productCode = Products.productCode
LEFT JOIN Orders ON Orders.orderNumber = OrderDetails.orderNumber
JOIN Customers ON Customers.customerNumber = Orders.customerNumber
GROUP BY Products.productName
;
""",conn)
```

##### 5.Donner le nombre de commande pour chaque pays du client, ainsi que le montant total des commandes et le montant total payé : on veut conserver les clients n’ayant jamais commandé dans le résultat final ;

```python
pandas.read_sql_query(
"""
SELECT c.country,
       COUNT(DISTINCT o.orderNumber) AS Nombre_commande,
       SUM(od.quantityOrdered * od.priceEach) AS montant_total_commande,
       SUM(p.amount) AS montant_total_payé
FROM Customers c
LEFT JOIN Orders o ON c.customerNumber = o.customerNumber
LEFT JOIN OrderDetails od ON o.orderNumber = od.orderNumber
LEFT JOIN Payments p ON c.customerNumber = p.customerNumber
GROUP BY c.country
;
""",
conn)


                            
```

##### 6.On veut la table de contigence du nombre de commande entre la ligne de produits et le pays du client ;


```python
pandas.read_sql_query(
"""
SELECT   Products.productName as "Produits", Customers.country as "Pays du client", COUNT(Orders.orderNumber) as "Nombre de commande"
FROM Products 
JOIN OrderDetails on OrderDetails.productCode = Products.productCode
JOIN Orders ON Orders.orderNumber = OrderDetails.orderNumber
JOIN Customers ON Customers.customerNumber = Orders.customerNumber
GROUP BY Customers.country 
;
""",conn)

```

##### 7.On veut la même table croisant la ligne de produits et le pays du client, mais avec le montant total payé dans chaque cellule ;



```python
pandas.read_sql_query(
"""
SELECT 
    Products.productLine AS "Ligne de produits", 
    Customers.country AS "Pays du client", 
    SUM(Payments.amount) AS "Montant total payé"
FROM Products
JOIN OrderDetails ON OrderDetails.productCode = Products.productCode
JOIN Orders ON Orders.orderNumber = OrderDetails.orderNumber
JOIN Customers ON Customers.customerNumber = Orders.customerNumber
JOIN Payments ON Payments.customerNumber = Customers.customerNumber
GROUP BY Products.productLine, Customers.country
ORDER BY Products.productLine, Customers.country
;
""",
conn)
```

##### 8.Donner les 10 produits pour lesquels la marge moyenne est la plus importante (cf buyPrice et priceEach) ;


```python
pandas.read_sql_query(
"""
SELECT  Products.productName, Products.buyPrice, OrderDetails.priceEach, AVG(OrderDetails.priceEach-Products.buyPrice)as "marge moyenne"
FROM Products
JOIN OrderDetails ON OrderDetails.productCode = Products.productCode
GROUP BY Products.productName
ORDER BY  "marge moyenne" desc
LIMIT 10
;
""",conn)
```

##### 9.Lister les produits (avec le nom et le code du client) qui ont été vendus à perte :##### 
Si un produit a été dans cette situation plusieurs fois, il doit apparaître plusieurs fois 
##### 
Une vente à perte arrive quand le prix de vente est inférieur au prix d’achat ;

```python
pandas.read_sql_query(
"""
SELECT 
p.productName,
p.productCode
FROM Products p
JOIN OrderDetails od on od.productCode = p.productCode
WHERE priceEach < buyPrice 
GROUP BY p.productLine
;
""",
conn)
```
